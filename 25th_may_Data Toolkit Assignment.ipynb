```python
=>> The difference in np.array, np.asarray and np.asanyarray
   1=>> np.array = Always creates a new array, copying data unless specified otherwise.
    - Always creates a new array, copying the data unless copy=False is explicitly specified and the input is already an array of the correct type.
    - Use when you want to ensure you have a NumPy array and may need to specify a different data type or require a copy of the input data.
  

    code = import numpy as np

         # Creating a new array from a list
          a = np.array([1, 2, 3])
            print(a)

         # Creating a new array from an existing array (always copies unless specified otherwise)
          b = np.array(a)
            print(b)

  2=>> np.asarray: Converts to a NumPy array without copying if the input is already an array.
    - Converts the input to a NumPy array, but does not copy the data if the input is already a NumPy array (or array-like) of the correct type.
    - Creates an array only if the input is not already a NumPy array. If the input is already a NumPy array,
      it returns the input as is, without copying.
    - Use when you want to ensure you have a NumPy array without incurring the cost of copying data if the input is already an array

    code = # Creating an array from a list
           c = np.asarray([1, 2, 3])
             print(c)

           # No copy is made if the input is already a NumPy array
        d = np.asarray(c)
             print(d)

    
    3=>> np.asanyarray: Similar to np.asarray, but preserves subclasses of ndarray.
    - Similar to np.asarray, but passes through any subclasses of ndarray.  This means that if the input is an instance of a subclass of ndarray, 
      such as matrix or masked arrays, the output will be of the same subclass
    - Converts the input to a NumPy array or an array of the same subclass, without copying data if the input is already an array or of the correct subclass type.
    -  Use when you want to ensure you have an array or maintain the subclass type of the input if it is already a subclass of ndarray.

    code = # Creating an array from a list
              e = np.asanyarray([1, 2, 3])
               print(e)

           # Maintains subclass if input is a subclass of ndarray
            matrix = np.matrix([[1, 2], [3, 4]])
               f = np.asanyarray(matrix)
                print(f)
                print(type(f))  # Will be <class 'numpy.matrix'>

         
```

```python
- The difference between Deep copy and shallow copy
  1=>> Deep Copy = A deep copy creates a new object and recursively copies all objects found in the original.
                   It creates entirely new objects for all nested objects
                - Copies the structure and all nested objects, creating independent copies.
                - Changes to the nested objects in the copied structure will not affect the original structure and vice versa.
                
CODE = 
import copy

original_list = [[1, 2, 3], [4, 5, 6]]
deep_copy_list = copy.deepcopy(original_list)

print("Original List:", original_list)
print("Deep Copy List:", deep_copy_list)

# Modify the first element of the first nested list
deep_copy_list[0][0] = 100

print("After Modification")
print("Original List:", original_list)
print("Deep Copy List:", deep_copy_list)


 2=>> Shallow Copy = A shallow copy creates a new object, but does not create copies of the objects that the original object references. 
                    Instead, it inserts references into the new object that point to the same objects that the references in the original object point to.
                   - Copies the structure of the original object, but not the nested objects within it.
                   - Changes to the nested objects in the copied structure will reflect in the original structure and vice versa.

    CODE = 
import copy

original_list = [[1, 2, 3], [4, 5, 6]]
shallow_copy_list = copy.copy(original_list)

print("Original List:", original_list)
print("Shallow Copy List:", shallow_copy_list)

# Modify the first element of the first nested list
shallow_copy_list[0][0] = 100

print("After Modification")
print("Original List:", original_list)
print("Shallow Copy List:", shallow_copy_list)
```

```python
14) Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and 
Perform the following tasks using Matplotlib:
    
a) Plot the 'Temerature' and 'Humidity' on the same lot with different y-axes (left y-axis for 'Temerature' and 
right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the lot as 'Temerature and Humidity Over Time'.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a time-series dataset
date_range = pd.date_range(start='2023-01-01', periods=100, freq='D')
temperature = np.random.uniform(low=-10, high=35, size=100)  # Random temperatures between -10 and 35 degrees
humidity = np.random.uniform(low=20, high=100, size=100)  # Random humidity between 20% and 100%

# Create a DataFrame
data = {
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
}
df = pd.DataFrame(data)

# a) Plot 'Temperature' and 'Humidity' on the same plot with different y-axes
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.plot(df['Date'], df['Temperature'], color='tab:red', label='Temperature')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.plot(df['Date'], df['Humidity'], color='tab:blue', label='Humidity')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# b) Label the x-axis as 'Date'
# This is already done in ax1.set_xlabel('Date')

# c) Set the title of the plot as 'Temperature and Humidity Over Time'
plt.title('Temperature and Humidity Over Time')

# Adding legends
fig.legend(loc='upper left', bbox_to_anchor=(0.1,0.9))

# Show the plot
plt.show()
```

```python
15) Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following 
tasks using Matplotlib :

a) Plot a histogram of the data with 30 bins.

b) Overlay a line lot reresenting the normal distribution's robability density function (PDF).

c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

d) Set the title of the lot as 'Histogram with PDF Overlay
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create a NumPy array with 1000 samples from a normal distribution
data = np.random.randn(1000)

# a) Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# b) Overlay a line plot representing the normal distribution's probability density function (PDF)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, data.mean(), data.std())
plt.plot(x, p, 'k', linewidth=2, label='Normal PDF')

# c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# d) Set the title of the plot as 'Histogram with PDF Overlay'
plt.title('Histogram with PDF Overlay')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
17) Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the 
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create two random arrays
np.random.seed(0)  # For reproducibility
x = np.random.randn(100)
y = np.random.randn(100)

# Determine the quadrant for each point
def determine_quadrant(x, y):
    if x > 0 and y > 0:
        return 'Quadrant I'
    elif x < 0 and y > 0:
        return 'Quadrant II'
    elif x < 0 and y < 0:
        return 'Quadrant III'
    else:
        return 'Quadrant IV'

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y})
df['Quadrant'] = df.apply(lambda row: determine_quadrant(row['x'], row['y']), axis=1)

# Create a scatter plot using Seaborn
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='deep', s=100)

# Add legend
plt.legend(title='Quadrant')

# Label the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set the title
plt.title('Quadrant-wise Scatter Plot')

# Show the plot
plt.show()
```

```python
19) Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their 
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical 
Bar Chart'
```

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource, HoverTool

# Enable output in the notebook
output_notebook()

# Generate random categorical data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(1, 101, size=len(categories))

# Create a DataFrame
data = pd.DataFrame({'categories': categories, 'values': values})

# Create a ColumnDataSource
source = ColumnDataSource(data)

# Define a color map based on the values
colors = factor_cmap('categories', palette=['#f46d43', '#fdae61', '#fee08b', '#d73027', '#a50026'], factors=categories)

# Create a new plot
p = figure(x_range=categories, title="Random Categorical Bar Chart", x_axis_label='Categories', y_axis_label='Values', tooltips=[("Value", "@values")])

# Add bars to the plot
p.vbar(x='categories', top='values', width=0.9, source=source, legend_field='categories', line_color='white', fill_color=colors)

# Customize the plot
p.legend.title = "Categories"
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Show the plot
show(p)
```
